```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from matplotlib.dates import date2num
import seaborn as sns
```

```python
# read data
df = pd.read_excel('Madison crashes 1998 - July 2019.xlsx')
```

```python
# change type to datetime
df.ACCDDATE = pd.to_datetime(df.ACCDDATE)
```

```python
# set mask
start_date = pd.to_datetime("01/01/2009")
end_date = pd.to_datetime("12/31/2019")
mask = (df.ACCDDATE > start_date) & (df.ACCDDATE <= end_date)
```

```python
# get data within date range
df = df.loc[mask]
export_csv = df.to_csv(r'/Users/ark.kuo/Project/DS/V0/Madison_crash_2009_2019.csv', index = None, header=True) #Don't forget to add '.csv' at the end of the path
```
